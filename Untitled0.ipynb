```python
#@title 백준 4344
number=int(input())
for i in range(number):
    count=0
    num=input().split()
    scores=list(map(float,num[1:]))
    average=sum(scores)/len(scores)
    for temp in scores:
        if temp>average:
            count+=1
    print('{:.3f}%'.format(count/len(scores)*100))
```

```python
#@title 백준 8958
number=int(input())
for i in range(number):
    count=0
    sum=0
    data=list(input())
    for temp in data:
        if temp=='O':
            count+=1
            sum+=count
        elif temp=='X':
            count=0
    print(sum)
```
